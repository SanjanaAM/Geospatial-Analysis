```python
#flood

import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
!pip install scikit-image  # Install scikit-image for resizing

from skimage.transform import resize

# Load two images (before and after flood)
with rasterio.open('/content/valenciaflooding_pre_image.tiff') as src:
    before = src.read(1)
    before_meta = src.meta  # Store metadata for later

with rasterio.open('/content/valenciaflooding_post_image.tiff') as src:
    after = src.read(1)
    after_meta = src.meta

# Resize 'before' image to match 'after' image dimensions
before_resized = resize(before, after.shape, order=0, preserve_range=True)

# Now 'before_resized' and 'after' have the same shape
change = after - before_resized

# Visualize the change with color map
plt.imshow(change, cmap='RdYlGn')
cbar = plt.colorbar(label='Change')

# Set custom labels for the colorbar
# Defining the range for custom labels
vmin = change.min()
vmax = change.max()

# Custom ticks based on the values in the 'change' array
ticks = np.linspace(vmin, vmax, 5)  # 5 intervals for 6 labels
cbar.set_ticks(ticks)  # Set ticks to match the custom range

# Set custom labels to represent different flood severity categories
cbar.ax.set_yticklabels([
    'No Flooding',         # First tick (no flooding)
    'Minor Flooding',      # Second tick (slight flooding)
    'Flooded',             # Third tick (moderate flooding)
    'Severe Flooding',     # Fourth tick (severe flooding)

    'Extreme Flooding',    # Fifth tick (maximum flooding)
])

# Add a title for the plot
plt.title('Flood Detection')

# Show the plot
plt.show()
```

```python
# Hurricane
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from skimage.transform import resize # Import resize for image resizing

# Load two images (before and after disaster)
with rasterio.open('/content/your_image.tif') as src:
    before = src.read(1)

with rasterio.open('/content/tropical 2 fin.tif') as src:
    after = src.read(1)

# Resize 'before' image to match 'after' image dimensions if they are different
if before.shape != after.shape:
    before = resize(before, after.shape, order=0, preserve_range=True)

# Now 'before' and 'after' have the same shape (or are broadcastable)
change = after - before

# Visualize the change with color map
plt.imshow(change, cmap='RdYlGn')
cbar = plt.colorbar(label='Change')

# Set custom labels for the colorbar
# Defining the range for custom labels
vmin = change.min()
vmax = change.max()

# Custom ticks based on the values in the 'change' array
ticks = np.linspace(vmin, vmax, 5)  # 5 intervals for 6 labels
cbar.set_ticks(ticks)  # Set ticks to match the custom range

# Set custom labels to represent different change categories
cbar.ax.set_yticklabels([
    'Moderate Damage',   # First tick (minimum damage)
    'Minor Damage',      # Second tick
    'No Change',         # Third tick
    'Minor Recovery',    # Fourth tick (centered around 0)

    'Moderate Recovery', # Fifth tick
   # 'Major Recovery'     # Sixth tick (maximum recovery)
])

# Add a title for the plot
plt.title('Hurricane Detection')

# Show the plot
plt.show()
```
